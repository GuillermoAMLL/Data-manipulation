Seleccionar valores específicos de un DataFrame o Serie de pandas para trabajar es un paso implícito en casi cualquier operación de datos que ejecute, por lo que una de las primeras cosas que debe aprender al trabajar con datos en Python es cómo seleccionar los datos. puntos relevantes para usted de forma rápida y eficaz.

```python
#Indexación basada en la posición númerica del dato iloc
#Selección de la primrea fila del set.
wine.iloc[0]
```

Tanto loc como iloc son la primera fila y la segunda columna. Esto es lo opuesto a lo que hacemos en Python nativo, que es la primera columna y la segunda fila.

Esto significa que es un poco más fácil recuperar filas y un poco más difícil recuperar columnas. Para obtener una columna con iloc, podemos hacer lo siguiente:

```python
#Selección columna

wine.iloc[:,0]
```

Por sí solo, el operador :, que también proviene del Python nativo, significa "todo". Sin embargo, cuando se combina con otros selectores, se puede utilizar para indicar un rango de valores. Por ejemplo, para seleccionar la columna de país solo de la primera, segunda y tercera fila, haríamos:

```python
wine.iloc [:3,0]
```

```python
#Selección de segunda y tercera entrada de dato

wine.iloc[1:3,0]
```

```python
#Código como lista

wine.iloc[[0,1,2],0]
```

```python
#Números negativos utilizados como selección inversa, desde el final
#Los últimos 5 registros del data set wines
wine.iloc[-5:]
```

```python
#Selección de columna completa country
wine.loc[:,'country']
```

Al elegir o realizar la transición entre loc e iloc, hay un problema que vale la pena tener en cuenta: los dos métodos utilizan esquemas de indexación ligeramente diferentes.

iloc utiliza el esquema de indexación stdlib de Python, donde se incluye el primer elemento del rango y se excluye el último. Entonces 0:10 seleccionará las entradas 0,...,9. loc, mientras tanto, indexa inclusive. Entonces 0:10 seleccionará las entradas 0,...,10.

### Selección condicional
Por ejemplo, supongamos que estamos interesados específicamente en vinos mejores que el promedio producidos en Italia.

```python
wine.country=='Italy'
```

```python
#Selección condicional para el dataset solo para vinos de Italia (Tabla completa)
wine.loc[wine.country=='Italy']
```

Ahora necesitamos saber que vinos son mejores en promedio si su origen es italiano, considerando la variable review (1-100) puntos, en este caso el caso de mejores vinos esta en el tramo de (80 a 100) en reviews, es decir al menos 90 puntos.

```python
#Selección con dos condicionales; país y puntos
wine.loc[(wine.country=='Italy') & (wine.points>=90) ]
```

Supongamos que compramos cualquier vino elaborado en Italia o que tenga una calificación superior a la media. Para esto utilizamos un encadenar (|)

```python
wine.loc[(wine.country=='Italy') & (wine.points >=90)]
```

Pandas viene con algunos selectores condicionales integrados, dos de los cuales destacaremos aquí.

El primero es isin. isin le permite seleccionar datos cuyo valor "está en" una lista de valores. Por ejemplo, así es como podemos usarlo para seleccionar vinos solo de Italia o Francia:

```python
#Filtrar datos con isin para encontrar vinos de Italia y Francia
wine.loc[wine.country.isin(['Italy', 'France'])]
```

El segundo es nulo (y su compañero no nulo). Estos métodos le permiten resaltar valores que están (o no) vacíos (NaN). Por ejemplo, para filtrar vinos sin etiqueta de precio en el conjunto de datos, esto es lo que haríamos:

```python
#Utilizando función isnull (es nulo) y notnull (no nulo)
#Filtrar vinos sin etiqueta de precio, es decir valores nulos de la columna precios
wine.loc [wine.price.notnull()]
```

#### Mapas

Un mapa es un término, tomado de las matemáticas, para una función que toma un conjunto de valores y los "mapea" a otro conjunto de valores. En la ciencia de datos, a menudo tenemos la necesidad de crear nuevas representaciones a partir de datos existentes, o de transformar datos del formato en el que se encuentran ahora al formato en el que queremos que estén más adelante. Los mapas son los que se encargan de este trabajo, lo que los hace extremadamente importantes para realizar el trabajo.

Hay dos métodos de mapeo que utilizará con frecuencia.

map() es el primero y un poco más simple. Por ejemplo, supongamos que quisiéramos resignificar las puntuaciones que recibieron los vinos a 0. Podemos hacer esto de la siguiente manera:

```python
wine_points_mean = wine.points.mean()

wine.points.map(lambda p:p - wine_points_mean)
```

La función que pasa a map() debe esperar un valor único de la Serie (un valor en puntos, en el ejemplo anterior) y devolver una versión transformada de ese valor. map() devuelve una nueva Serie donde todos los valores han sido transformados por su función.

apply() es el método equivalente si queremos transformar un DataFrame completo llamando a un método personalizado en cada fila.

```python
#Creación de función

def remean_points(row):
  row.points = row.points - wine_points_mean
  return row

wine.apply(remean_points, axis= 'columns')
```

Si hubiéramos llamado a reviews.apply() con axis='index', entonces en lugar de pasar una función para transformar cada fila, necesitaríamos proporcionar una función para transformar cada columna.

Tenga en cuenta que map() y apply() devuelven series y marcos de datos nuevos y transformados, respectivamente. No modifican los datos originales a los que son convocados. Si miramos la primera fila de reseñas, podemos ver que todavía tiene su valor de puntos original.

Hay un número limitado de palabras que puedes utilizar al describir una botella de vino. ¿Es más probable que un vino sea "tropical" o "afrutado"? Cree una serie descriptor_counts contando cuántas veces aparece cada una de estas dos palabras en la columna de descripción del conjunto de datos. (Para simplificar, ignoremos las versiones en mayúscula de estas palabras).

```python
n_trop = wine.description.map(lambda desc: "tropical" in desc).sum()

n_fruity = wine.description.map(lambda x: "fruity" in x).sum()

descriptor_counts = pd.Series([n_trop, n_fruity], index = ['tropical', 'fruity'])

descriptor_counts
```

Nos gustaría publicar estas reseñas de vinos en nuestro sitio web, pero un   sistema de calificación que oscila entre 80 y 100 puntos es demasiado difícil de entender; nos gustaría traducirlas en simples calificaciones de estrellas. Una puntuación de 95 o más cuenta como 3 estrellas, una puntuación de al menos 85 pero menos de 95 es 2 estrellas. Cualquier otra puntuación es 1 estrella.

Además, la Asociación Canadiense de Viñeros compró muchos anuncios en el sitio, por lo que cualquier vino de Canadá debería obtener automáticamente 3 estrellas, independientemente de los puntos.

Cree una serie star_ratings con la cantidad de estrellas correspondientes a cada reseña en el conjunto de datos.

```python
def stars (row):
    if row.country =='Canada':
        return 3
    elif row.points >= 95:
        return 3
    elif row.points >=85:
        return 2
    else:
        return 1

star_ratings =reviews.apply(stars, axis='columns')
star_ratings
```
